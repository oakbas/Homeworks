## Two Event is Independent or Not?

Join Probability of Two Event: 

\begin{align*}
p(X_{1},X_{2}) = \left\{ \begin{array}{cc} p(X_{1}) \; p(X_{2}) & m=0 \\
p(X_{1}) \; p(X_{2} | p(X_{1}) & m=1\end{array} \right.
\end{align*}

\begin{align*}
p(X|m) = \left\{ \begin{array}{cc} \int \int d\pi_{1} d\pi_{2} p(X, \pi_{1}, \pi_{2})& m=0 \\
\int \int \int d\pi_{1} d\theta_{0,1} d\theta_{1,1} p(X, \pi_{1}, \theta_{0,1}, \theta_{1,1}) & m=1\end{array} \right.
\end{align*}

We know that probability of all the parameters are Beta distribution:

\begin{align*}
\pi_{1} & \sim & \mathcal{B}(\alpha, \beta) \\
\pi_{2} & \sim & \mathcal{B}(\alpha, \beta) \\
\theta_{0,1} & \sim & \mathcal{B}(\alpha, \beta) \\
\theta_{1,1} & \sim & \mathcal{B}(\alpha, \beta) \\
\end{align*}


#### For m = 0
\begin{eqnarray}
p(X,\pi_{1}, \pi_{2}) &=& p(x_{1} | \pi_{1}) . p(x_{2} | \pi_{2}) . p(\pi_{1}) . p(\pi_{2})\\
p(X_{1}, X_{2}, \pi_{1}, \pi_{2}) &=& \prod_{i=1}^N \left(p(x_{1}^{(i)} | \pi_{1} .  p(\pi_{1})\right) \prod_{i=1}^N \left(p(x_{2}^{(i)} | \pi_{2} .  p(\pi_{2})\right) \\
\end{eqnarray}

\begin{eqnarray}
p(X_{1},X_{2}) &=& \int \int d\pi_{1} d\pi_{2} p(x_{1} | \pi_{1}) . p(x_{2} | \pi_{2}) . p(\pi_{1}) . p(\pi_{2})\\
&=& \left(\int d\pi_{1} \prod_{i=1}^N \pi_{1}^{x_{1}^{i}} {(1-\pi_{1})}^{1-x_{1}^{i}} \mathcal{B}(\alpha , \beta)\right ) .
\left (\int d\pi_{2} \prod_{i=1}^N \pi_{2}^{x_{2}^{i}} {(1-\pi_{2})}^{1-x_{2}^{i}} \mathcal{B}(\alpha , \beta) \right)\\
&=& \int d\pi_{1} \left(\prod_{i=1}^N \pi_{1}^{x_{1}^{i}} {(1-\pi_{1})}^{1-x_{1}^{i}}\right ).\pi_{1}^{\alpha-1} {(1-\pi_{1})}^{\beta-1}. 
\frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} 
\int d\pi_{2} \left (\prod_{i=1}^N \pi_{2}^{x_{2}^{i}} {(1-\pi_{2})}^{1-x_{2}^{i}}\right ) \pi_{2}^{\alpha-1} {(1-\pi_{2})}^{\beta-1} .
\frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} \\
&=& \int d\pi_{1}\pi_{1}^{c_{10}+\alpha-1} . {(1-\pi_{1})}^{c_{00}+\beta-1} \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} \int d\pi_{2}\pi_{2}^{c_{11}+\alpha-1} . {(1-\pi_{2})}^{c_{10}+\beta-1} \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)}  \\
&=& \frac{\Gamma(\alpha + c_{10}) \Gamma(\beta + c_{00})}{\Gamma(c_{10}+ c_{00} + \alpha + \beta)} . \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)} .  
\frac{\Gamma(\alpha + c_{11}) \Gamma(\beta + c_{10})} {\Gamma(c_{11}+ c_{10} + \alpha + \beta)}. \frac{\Gamma(\alpha + \beta)}{\Gamma(\alpha) \Gamma(\beta)}  
\end{eqnarray}

#### For m = 1
